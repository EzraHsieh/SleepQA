Training Dataset

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split


RANDOM_SEED = 42
```

## Combine to obtain full 5000 triplets data

```python
# Combine training, validation, and test json datasets
train_data = pd.read_json('sleep-train.json')
dev_data = pd.read_json('sleep-dev.json')
test_data = pd.read_json('oracle/sleep-test.json')
display(train_data)

data = pd.concat([train_data, dev_data, test_data], ignore_index=True)
data.to_json('sleep-full.json', orient='records', lines=True)
```

```python
# Create subset of full data to test how sample size impact model performance
# 100 triplets
data_100 = data.sample(n=100, random_state=RANDOM_SEED)
data_100.to_json('sleep-100.json', orient='records', lines=True)
# 200 triplets
data_200 = data.sample(n=200, random_state=RANDOM_SEED)
data_200.to_json('sleep-200.json', orient='records', lines=True)
# 500 triplets
data_500 = data.sample(n=500, random_state=RANDOM_SEED)
data_500.to_json('sleep-500.json', orient='records', lines=True)
# 1000 triplets
data_1000 = data.sample(n=1000, random_state=RANDOM_SEED)
data_1000.to_json('sleep-1000.json', orient='records', lines=True)
# 2500 triplets
data_2500 = data.sample(n=2500, random_state=RANDOM_SEED)
data_2500.to_json('sleep-2500.json', orient='records', lines=True)
```

```python
# Train, dev, test split for each subset data randomly

def split_and_save(data, prefix):
    train, temp = train_test_split(data, test_size=0.2, random_state=RANDOM_SEED)
    dev, test = train_test_split(temp, test_size=0.5, random_state=RANDOM_SEED)
    
    train.to_json(f'{prefix}-train.json', orient='records', lines=False)
    dev.to_json(f'{prefix}-dev.json', orient='records', lines=False)
    test.to_json(f'{prefix}-test.json', orient='records', lines=False)

split_and_save(data_100, 'sleep-100')
split_and_save(data_200, 'sleep-200')
split_and_save(data_500, 'sleep-500')
split_and_save(data_1000, 'sleep-1000')
split_and_save(data_2500, 'sleep-2500')
    
```
